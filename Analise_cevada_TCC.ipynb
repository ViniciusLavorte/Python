```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
base = pd.read_csv('Resultados_TCC2 - eval.csv')
```

```python
base.head()
```

```python
Precisao = ['Rede','Precisao_1','Precisao_2','Precisao_3','Precisao_4','Precisao_5','Precisao_6','Precisao_7','Precisao_8', 'Precisao_final']
Tempo = ['Rede','Tempo_1','Tempo_2','Tempo_3','Tempo_4','Tempo_5','Tempo_6','Tempo_7','Tempo_8','Tempo_final']
bp = ['P1','P2','P3','P4','P5','P6','P7','P8','P_Final']
```

```python
b = base.loc[(base.Amostra_Dados == 'cevada') & (base.Tipo == 'Normal'), Precisao]
b = b.T
b.columns = b.iloc[0,:]
b = b.drop(['Rede'])
b.index = bp

t = base.loc[(base.Amostra_Dados == 'cevada') & (base.Tipo == 'Normal'), Tempo]
t = t.T
t.columns = t.iloc[0,:]
t = t.drop(['Rede'])
t.index = bp
```

```python
plt.plot(b.iloc[:-1,:],marker ='o')
plt.title('Accuracy: Larvae Normal')
plt.legend(b.columns,loc='center left', bbox_to_anchor=(1, 0.7))
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.show()

plt.plot(t.iloc[:-1,:],marker ='o')
plt.title('Accuracy: Larvae Normal')
plt.legend(t.columns,loc='center left', bbox_to_anchor=(1, 0.7))
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.show()

```
