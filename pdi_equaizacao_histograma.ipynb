### Processamento de Imagem: Equalização de Histograma

```python
import numpy as np
from matplotlib import pyplot as plt
import math as m
```

```python
#Níveis de cinza (k)
nc = np.array([0,1,2,3,4,5,6,7])
#Número de pixels (nk)
npi = np.array([120, 200, 350, 400, 100, 80, 50, 30])
```

##### Plotar histogram original

```python
plt.bar(nc, npi) # bar(coordenadas(tipo) ,quantidade )
plt.title('Histograma original')
plt.ylabel('Qtd de cinzas')
plt.show()
```

```python
#Função para normalizar os dados
def normalize(sem_norm):
    norm =[]
    for c in sem_norm:
        norm.append(np.divide( c,float(np.sum(sem_norm))).round(3))
    return norm 
```

```python
#normalizando os dados
dados_normal = normalize(npi)
soma = []
tot = 0
nivel_cinza = []
for k in dados_normal:
    #somatoria
    tot = tot + k
    soma.append(tot)
    #nivel cinza proximo
    nivel_cinza.append(round(np.multiply(len(dados_normal)-1,tot))) 
nivel_cinza
```

```python
plt.bar(nivel_cinza, npi) # bar(coordenadas(tipo) ,quantidade )
plt.title('Histograma original')
plt.ylabel('Qtd de cinzas')
plt.show()
```
