```python
import numpy as np
import pandas as pd
#import sqlalchemy as sqla
import matplotlib.pyplot as plt
base = pd.read_csv('Resultado_processado_TCC2_eval.csv',delimiter=';')
```

```python
base
```

```python
base.columns
```

```python
Precisao = ['Pasta','Precisao_1','Precisao_2','Precisao_3','Precisao_4','Precisao_5','Precisao_6','Precisao_7','Precisao_8', 'Precisao_final']
Tempo = ['Pasta','Tempo_1','Tempo_2','Tempo_3','Tempo_4','Tempo_5','Tempo_6','Tempo_7','Tempo_8','Tempo_final']
```

## Inception_v3 and Larvae

```python
b = base.loc[(base.Amostra_Dados == 'larvae') & (base.Rede == 'Inception_V3'),Precisao]
bt = b.T
bt.columns = bt.iloc[0]
bt = bt.drop('Pasta')
bt_row = ['P1','P2','P3','P4','P5','P6','P7','P8','P_Final']
bt.index = bt_row
bt
```

```python
# Precisão Inception_v3 e Larvae
b = base.loc[(base.Amostra_Dados == 'larvae') & ((base.Rede == 'Resnet_v2')|(base.Rede == 'Nasnet')|(base.Rede == 'Inception_V3')),Precisao]
bt = b.T
bt.columns = bt.iloc[0]
bt = bt.drop(['Pasta'])
bt_row = ['P1','P2','P3','P4','P5','P6','P7','P8','P_Final']
bt.index = bt_row
#bt = bt.fillna(bt.iloc[1:,].mean())


# Tempo Inception_v3 e Larvae
c = base.loc[(base.Amostra_Dados == 'larvae') & ((base.Rede == 'Resnet_v2')|(base.Rede == 'Nasnet')|(base.Rede == 'Inception_V3')),Tempo]
ct = c.T
ct.columns = ct.iloc[0]
ct = ct.drop(['Pasta'])
ct_row = ['T1','T2','T3','T4','T5','T6','T7','T8','T_Final']
ct.index =  ct_row
#ct = ct.fillna(ct.iloc[-1,:])
ct
```

## Plotando Inception_v3 and Larvae

```python
# Plotando as saídas de Accuracia e tempo
#plt.figure(figsize=(9, 3))

plt.plot(bt.iloc[:-1,:],marker ='o')
plt.title('Accuracy: Larvae')
plt.legend(bt.columns,loc='center left', bbox_to_anchor=(1, 0.7))
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.show()


plt.plot(ct.iloc[:-1,:],marker ='o')
plt.title('Tempo: Larvae')
plt.legend(bt.columns,loc='center left', bbox_to_anchor=(1, 0.7))
plt.ylabel('Tempo')
plt.xlabel('Epoch')
plt.show()
```

## Inception_v3 and Cevada

```python
# Precisão Inception_v3 e Cevada
b = base.loc[(base.Amostra_Dados == 'cevada') & ((base.Rede == 'Resnet_v2')|(base.Rede == 'Nasnet')|(base.Rede == 'Inception_V3')),Precisao]
bt = b.T
bt.columns = bt.iloc[0]
bt = bt.drop(['Pasta'])
bt_row = ['P1','P2','P3','P4','P5','P6','P7','P8','P_Final']
bt.index = bt_row
#bt = bt.fillna(bt.iloc[1:,].mean())


# Tempo Inception_v3 e Cevada
c = base.loc[(base.Amostra_Dados == 'cevada') & ((base.Rede == 'Resnet_v2')|(base.Rede == 'Nasnet')|(base.Rede == 'Inception_V3')),Tempo]
ct = c.T
ct.columns = ct.iloc[0]
ct = ct.drop(['Pasta'])
ct_row = ['T1','T2','T3','T4','T5','T6','T7','T8','T_Final']
ct.index =  ct_row
#ct = ct.fillna(ct.iloc[-1,:])
ct
```

## Inception_v3 and Mammoset

```python
# Precisão Inception_v3 e Mammoset
b = base.loc[(base.Amostra_Dados == 'mammoset') & ((base.Rede == 'Resnet_v2')|(base.Rede == 'Nasnet')|(base.Rede == 'Inception_V3')),Precisao]
bt = b.T
bt.columns = bt.iloc[0]
bt = bt.drop(['Pasta'])
bt_row = ['P1','P2','P3','P4','P5','P6','P7','P8','P_Final']
bt.index = bt_row
#bt = bt.fillna(bt.iloc[1:,].mean())


# Tempo Inception_v3 e Mammoset
c = base.loc[(base.Amostra_Dados == 'mammoset') & ((base.Rede == 'Resnet_v2')|(base.Rede == 'Nasnet')|(base.Rede == 'Inception_V3')),Tempo]
ct = c.T
ct.columns = ct.iloc[0]
ct = ct.drop(['Pasta'])
ct_row = ['T1','T2','T3','T4','T5','T6','T7','T8','T_Final']
ct.index =  ct_row
ct = ct.fillna(ct.iloc[-1,:])
ct
```
