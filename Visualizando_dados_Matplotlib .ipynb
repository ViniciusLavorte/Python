### Gráfico de Barra, Histograma, Linhas e Dispersão

Importando biblioteca matplotlib

```python
from matplotlib import pyplot as plt
```

### Cria um gráfico de linha, anos no eixo x, gdp no eixo y

Criando variavel

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
plt.plot(years, gdp, color='green', marker='o', linestyle='dashdot')
plt.title('GDP Nominal')
plt.ylabel('Bilhões de $')
plt.xlabel('Anos')
plt.show()
```

### Criando um gráfico de Barra, é uma boa escolha quando você quer mostrar como algumas quantidades variam entre um conjunto particular de itens

```python
#Criando variaveis
movie = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
print(list(enumerate(movie)))
```

```python
# enumerate(), keep a count of iterations
# when you don't want to save the return of functions we use '_'
xs = [i + 0.1 for i, _ in enumerate(movie)] # xs vai ser utilizado como coordenadas no plano 'x' para o gráfico de barra
# as barras do gráfico com as coordenadas x à esquerda[xs], alturas[num_oscars]
plt.bar(xs, num_oscars) # bar(coordenadas(tipo) ,quantidade )
plt.title('Meus filmes Favoritos')
plt.ylabel('# de Premiações')

# nomeia o eixo x com nomes de filmes na barra central
plt.xticks([i + 0.5 for i, _ in enumerate(movie)], movie) # coordenadas no ponto x
plt.show()
```

### Um gráfico de barras é utilizado também para criar gráficos de histogramas, a fim de explorar como os valores estão distribuídos

```python
import collections as cou
grades = [83,95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0] # criando lista
deciles = lambda x: x // (10 * 10) # lambda, cria uma função # //, divisor que retorna números int
histogr = cou.Counter(deciles(grade) for grade in grades)# Counter, conta número de objetos e retorna um dict
#histogr.values() # retorna os valores de um dict
plt.bar([x -4 for x in histogr.keys()], #move cada barra para esquerda em 4, histogr.values() retorna as chaves de uma dict
       histogr.values(), # retorna os valores de uma dict
        8 ) # da para cada barra a largura de 8
#plt.axis([-5, 105, 0, 5]) # eixo x de -5 até 105
                          # eixo y de 0 até 5
plt.xticks([10 * i for i in range(11)]) #rótulos do eixo x em 0, 10, 20 ..100
plt.show()
```

### Gráfico de Linhas

```python
variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_error = [x+y for x, y in zip(variance, bias_squared)] #zip, retorna uma lista de tuplas
xs = [i for i, _ in enumerate(variance)] # Salvar cada i(qtd de elem) do loop

#podemos fazer multiplas chamadas para o plt.plot
#para mostrar múltiplas séries no mesmo gráfico
plt.plot(xs, variance, 'g-', label= 'variance')#linha verde sólida
plt.plot(xs, bias_squared, 'r-.', label = 'bias^2')#lilnha de ponto tracejado vermelho
plt.plot(xs, total_error, 'b:', label = 'total error')#linha pontilhado azul
#porque atribuimos rotulos para cada série
#podemos obter uma legenda gratuita
#loc = 9 significa "top center"
plt.legend(loc=9)
plt.xlabel("complexidade do modelo")
plt.title("Compromisso entre polarização e variância")
plt.show()
```

### Gráficos de Dispersão

```python
friends = [70,65,72,63,71,64,60,64,67]
minutes = [175,170,205,120,220,130,105,145,190]
labels = 'a b c d e f g h i'
labels = labels.split()
plt.scatter(friends,minutes)# Gráfico de dispersão
# nomeia cada posição
for label, friend_count, minutes_count in zip(labels, friends, minutes):
    plt.annotate(label,
                xy = (friend_count, minutes_count), #coloca o rótulo com sua posição
                xytext=(5,-5), #posição xy que é colocado o texto
                textcoords='offset points')
plt.title("Minutos Diários vs. Números de Amigos")
```
